## Compare data

To implement content-based filtering, we focus on the movie "overview" field, which contains descriptions of each movie. Our aim is to compare these descriptions to suggest similar movies. For instance, if a user is currently viewing "Kung Fu Panda" on platforms like Amazon or Netflix, we want to recommend comparable movies. We achieve this by comparing the description of "Kung Fu Panda" with those of other movies. Those with the closest descriptions are then recommended. This involves comparing every description with every other, a task streamlined with the scikit-learn library.

```python
# Importing the TfidfVectorizer class from scikit-learn's feature_extraction.text module. 
# This class is used for converting a collection of raw text documents into a matrix of TF-IDF features. 
# TF-IDF stands for Term Frequency-Inverse Document Frequency, which is a numerical statistic that 
# reflects the importance of a word in a document relative to a collection of documents (corpus). 
# This process helps in normalizing the frequency of words by considering their importance not only 
# in the current document but also across the entire corpus. 
# TfidfVectorizer is particularly useful in natural language processing tasks such as document 
# classification, clustering, and information retrieval, where we need to represent text data in 
# a numerical format understandable by machine learning algorithms. 
# By converting text data into TF-IDF vectors, it helps in reducing the dimensionality of the feature space,
# making it more manageable for machine learning algorithms and improving computational efficiency.
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Sample collection of documents (corpus)
# Extracting the "overview" column from movies_df DataFrame and converting it into a list.
# If there are any missing values (na means not available) in the "overview" column, they are filled with an empty string.
corpus = movies_df["overview"].fillna("").to_list()
```

```python
# Initializing the TfidfVectorizer.
# The stop_words parameter is set to 'english', which means common English stopwords like 'a', 'the', 'in', etc., 
# will be ignored during vectorization.
tfidf_vectorizer = TfidfVectorizer(stop_words='english')
```

```python
# Fitting the vectorizer to the corpus, which learns the vocabulary and IDF (Inverse Document Frequency) statistics 
# from the corpus, and transforming the corpus into TF-IDF features.
tfidf_matrix = tfidf_vectorizer.fit_transform(corpus)
```

```python
# Converting the TF-IDF matrix (which is in sparse format) into a dense array representation using tfidf_matrix.toarray(), 
# and creating a DataFrame from it. 
# The column names are set to the feature names obtained from the TfidfVectorizer using tfidf_vectorizer.get_feature_names_out().
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf_vectorizer.get_feature_names_out())

# Display the created df
tfidf_df
```

## Similarity matrix

```python
# Importing the linear_kernel function from sklearn's metrics.pairwise module. 
# This function computes the dot product of the TF-IDF matrix with itself, resulting in a similarity matrix.
from sklearn.metrics.pairwise import linear_kernel 

# Computing the similarity matrix by calculating the dot product of the TF-IDF matrix with itself using linear_kernel.
similarity_matrix = linear_kernel(tfidf_matrix, tfidf_matrix)

# Display this similarity_matrix
similarity_matrix
```

### Create a function that, when provided with a specific movie and the desired number of similar movies, returns a list of those similar movies.

```python
def similar_movies(movie_title, nr_movies):
    """
    Retrieve a list of similar movies based on the provided movie title.

    Args:
        movie_title (str): The title of the movie for which similar movies are to be found.
        nr_movies (int): The maximum number of similar movies to return.

    Returns:
        list: A list of similar movie titles.

    """

    # Find the index of the provided movie title in the DataFrame
    movie_index = movies_df.loc[movies_df['title'] == movie_title].index.to_list()[0]
    # Retrieve similarity scores between the provided movie and all other movies
    similarity_scores = list(enumerate(similarity_matrix[movie_index]))
    # Sort the similarity scores in descending order
    similarity_scores_sorted = sorted(similarity_scores, key=lambda x: x[1], reverse=True)
    # Extract movie indices with positive similarity scores (excluding the provided movie itself)
    similar_movie_indices = [tpl[0] for tpl in similarity_scores_sorted[1:] if tpl[1] > 0]
    # Retrieve movie titles corresponding to the extracted indices
    similar_movie_titles = movies_df.iloc[similar_movie_indices]['title'].to_list()
    
    # If there are fewer similar movies than requested, return all similar movies
    if len(similar_movie_titles) < nr_movies:
        return similar_movie_titles
    # Otherwise, return only the specified number of similar movies
    return similar_movie_titles[:nr_movies]
```
